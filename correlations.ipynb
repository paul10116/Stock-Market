```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import  datetime
import seaborn as sns
import yfinance as yf

sns.set_style("darkgrid")
sns.set(font_scale=1.7) 
```

```python
simbols1 = 'TSM  ALB  APD  BHP  CE  CF  ECL  EMN  FMC  FNV  IFF  LYB  NUE  OLN  PPG  RGLD  RIO  RPM  RS  SCCO  SHW  SMG  SQM  STLD  VMC  WLK'
tickers_1 = yf.Tickers(simbols1)
simbol_data = tickers_1.history(start = '2016-1-1', actions= False, rounding=True )
```

```python
corr_matrix = simbol_data["Close"].corr().round(2)
positive_corr = corr_matrix > 0.4
negative_corr = corr_matrix < -0.4

plt.figure(figsize=(40,40))
plt.title("Correlation over 0.69")
sns.heatmap(corr_matrix[positive_corr], linewidths=2, linecolor="white", cmap='Greens', annot=True, annot_kws={"size":20}, cbar=False);
```

```python
plt.figure(figsize=(40,40))
plt.title("Correlation below -0.69")
sns.heatmap(corr_matrix[negative_corr], linewidths=2, linecolor="white", cmap='Reds', annot=True, annot_kws={"size":20}, cbar=False);
```
