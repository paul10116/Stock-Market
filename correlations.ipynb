```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import  datetime
import seaborn as sns
import yfinance as yf

sns.set_style("darkgrid")
sns.set(font_scale=1.7) 
```

```python
simbols1 = ' ABR   ACAD   ACRE   AEO   AIV   AKR   AMKR   ARI   AROC   AXL   BDN   BGS   CFFN   CHS   CIM   CLF   CNK   CNX   CPRX   CXW   DEI   DENN   DISH   DRH   EFC   ENLC   ESTE   ET   EVRI   EXEL   EXTR   FBP   FNB   FULT   GBDC   GT   GTN   HA   HBAN   HBI   HCSG   HR   HST   INN   IVZ   KAR   KEY   KMI   KOS   KRG   KTOS   KW   LXP   LXU   MAC   MARA   MAT   MD    MFA   MLKN   MODG   MPW   MRC   MTG   MWA   NEOG   NLY   NWBI   NWL   NYCB   OI   OII   ONB   PAA  PARA   PCG   PDM   PEB   PLAB   PRMW   PSEC   PTEN   RCM   RES   RIOT   RLJ   ROIC   RWT   SBH   SBRA   SCS   SHO   SIRI   SITC    SSRM   SXC   TEN   UMPQ   VGR   VIAV   VLY   VTRS   WETF   WTI   WWW  '
tickers_1 = yf.Tickers(simbols1)
simbol_data = tickers_1.history(start = '2010-1-1', actions= False, rounding=True )
```

```python
corr_matrix = simbol_data["Close"].corr().round(2)
positive_corr = corr_matrix > 0.69
negative_corr = corr_matrix < -0.69

plt.figure(figsize=(40,40))
plt.title("Correlation over 0.69")
sns.heatmap(corr_matrix[positive_corr], linewidths=2, linecolor="white", cmap='Greens', annot=True, annot_kws={"size":20}, cbar=False);
```

```python
plt.figure(figsize=(40,40))
plt.title("Correlation below -0.69")
sns.heatmap(corr_matrix[negative_corr], linewidths=2, linecolor="white", cmap='Reds', annot=True, annot_kws={"size":20}, cbar=False);
```
